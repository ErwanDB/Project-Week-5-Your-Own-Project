## Import libraries

```python
import pandas as pd
import numpy as np
import re
```

## Import spotify dataset

```python
spotify = pd.read_csv('../Datasets/data_spotify.csv')
spotify
```

## Clean and transform spotify dataset for merge

```python
x = spotify["artists"][0]
x
```

```python
re.sub(r"[^a-zA-Z0-9 ]","",x)
```

```python
def artist(x):
    x = re.sub(r"[^a-zA-Z0-9, ]","",x)
    return x
```

```python
spotify["artists"]=spotify["artists"].apply(artist)
```

```python
spotify["artists"] = spotify["artists"].apply(lambda x:x.casefold())
spotify["name"] = spotify["name"].apply(lambda x:x.casefold())
```

```python
def name(x):
    x = re.sub(r"[^a-zA-Z0-9 ]","",x)
    x = x.replace(" remastered","")
    x = x.strip(" ")
    return x
```

```python
spotify["name"] = spotify["name"].apply(name)
```

```python
spotify
```

```python
spotify.drop(["acousticness","duration_ms","explicit","id","liveness","loudness","popularity","speechiness"],axis=1,inplace=True)
```

```python
spotify
```

```python
spotify.drop_duplicates(inplace=True)
spotify
```

```python
df = pd.merge(left=chords, how='left', left_on=["Artists","Songs"],right=spotify, right_on=["artists","name"])
```

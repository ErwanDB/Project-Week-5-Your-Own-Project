# Data scraping

## Import libraries

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import urllib
import re
```

## List of pages

```python
pages = []

for i in range(1,21):
    url = f'https://www.ultimate-guitar.com/explore?&type[]=Chords&order=hitstotal_desc&page={i}'
    pages.append(url)
    
pages
```

<font color=blue>**Most popular of all times = most number of hits**

## Scrap Votes & Ratings

```python
Votes = []
Ratings = []

for a in range(len(pages)):
    url = pages[a]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    body = soup.find_all('div', class_="js-store")[0]
    body = str(body)
    
    votes = re.findall(r'votes&quot;:\d+',body)
    rating = re.findall(r'rating&quot;:\S{4}',body)
    
    for b,c in zip(votes,rating):
        b = b.replace('votes&quot;:','')
        Votes.append(b)
        c = c.replace('rating&quot;:','')
        Ratings.append(c)
```

```python
print(len(Votes))
print(len(Ratings))
```

## list of url for songs

```python
list_url = []

for j in range(len(pages)):    
    url = pages[j]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    body = soup.find_all('div', class_="js-store")[0]
    body = str(body)
    body = body.split('&quot;')

    for k in body:
        if 'https://tabs.ultimate-guitar.com/tab/' in k:
            list_url.append(k)

list_url
```

```python
len(list_url)
```

## Scrap Artists

```python
Artists = []

for a in range(len(list_url)):
    url = list_url[a]
    
    artist = re.findall(r'(?=tab/)\S+',url)
    artist = artist[0].split('/')
    artist = artist[-2]
    artist = artist.replace('-',' ')
    Artists.append(artist)
    
print(len(Artists))
print(Artists)
```

## Scrap Song title

```python
Songs = []

for a in range(len(list_url)):
    url = list_url[a]
    song = re.findall(r'(?=tab/)\S+',url)
    song = song[0].split('/')
    song = song[-1]
    song = re.sub('.(chords-\d+)$',"",song)
    song = song.replace('-',' ')
    Songs.append(song)
    
print(len(Songs))
print(Songs)
```

## Scrap Chords

```python
Chords = []

for a in range(len(list_url)):
    url = list_url[a]
    
    response=requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    body = soup.find_all('div', class_='js-store')[0]
    body = str(body)
    
    chords = re.findall(r'(?=\[ch])(.*?)(?=\[/ch\])',body)
    chords = [i.replace('[ch]','') for i in chords]
    #chords = [i.replace('[/ch]','') for i in chords]
    chords = list(set(chords))
    Chords.append(chords)

print(len(Chords))
```

```python
print(Chords)
```

## Build DataFrame

```python
df = pd.DataFrame({
    "Artists":Artists,
    "Songs":Songs,
    "Rating":Ratings,
    "Votes":Votes,
    "Chords":Chords
})

df
```

## Create Level

### List of chords

```python
lst_col = [i for i in df.Chords]
lst_col
```

```python
lst_chords = []

for i in lst_col:
    for l in i:
        lst_chords.append(l)
        
len(set(lst_chords))
```

<font color=blue>**448 unique chords**

### Frequence of chords

```python
chords = []
freq = []
for i in set(lst_chords):
    chords.append(i)
    freq.append(lst_chords.count(i))
chords_freq = pd.DataFrame({"freq":freq},index=chords)
chords_freq.sort_values(by=['freq'],ascending=False)
```

```python
chords_freq["freq"].nlargest(5)
```

```python
chords_freq["freq"].nlargest(15)
```

### Define Level

```python
# Top 5 most frequent chords. Easy chords without bar and neighbour frets
novice = ['Am','C','D','Em','G']

# 5 following more frequent chords. Player able to navigate fretboard playing bar chords with neighbour frets
intermediate = ['E','A','F','Bm','Dm']

# Complex chords with non neighbour frets and dampening strings
advanced = [i for i in list(set(lst_chords)) if i not in (novice + intermediate)]
```

```python
len(advanced)+len(intermediate)+len(novice)==len(list(set(lst_chords)))
```

```python
def level(lst):
    
    level = []
    
    for i in lst:
        if i in advanced:
            level.append("advanced")
        elif i in intermediate:
            level.append("intermediate")
        elif i in novice:
            level.append("novice")
            
    if 'advanced' in level:
        return "advanced"
    elif "intermediate" in level and 'advanced' not in level:
        return "intermediate"
    elif "novice" in level and "intermediate" not in level and 'advanced' not in level:
        return "novice"
```

### Add level column

```python
df["Level"]=df["Chords"].apply(level)
```

## Encode chords

```python
max_chords = max(list(map(lambda x:len(x),lst_col)))
max_chords
```

```python
columns = ['chord_{}'.format(i) for i in range(1,max_chords+1)]
df_chords = pd.DataFrame(lst_col, columns=columns)
df_chords
```

```python
df_chords.info()
```

```python
dummies = pd.get_dummies(df_chords[columns],prefix='', prefix_sep='')
dummies_combined_cols = dummies.groupby(dummies.columns, axis=1).agg(np.sum)
```

```python
len(list(dummies_combined_cols.columns))
```

```python
dummies_combined_cols
```

```python
chords_encod = pd.concat((df,dummies_combined_cols), axis=1)
chords_encod = chords_encod.drop(["Chords"],axis=1)
chords_encod
```

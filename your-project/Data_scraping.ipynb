# Data scraping

## Import libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib
import re
```

```python
pages = []

for i in range(1,21):
    url = f'https://www.ultimate-guitar.com/explore?&type[]=Chords&order=hitstotal_desc&page={i}'
    pages.append(url)
    
pages
```

```python
list_url = []

for j in range(len(pages)):    
    url = pages[j]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    body = soup.find_all('div', class_="js-store")[0]
    body = str(body)
    body = body.split('&quot;')

    for k in body:
        if 'https://tabs.ultimate-guitar.com/tab/' in k:
            list_url.append(k)

list_url
```

```python
len(list_url)
```

```python
Artists = []

for a in range(len(list_url)):
    url = list_url[a]
    
    artist = re.findall(r'(?=tab/)\S+',url)
    artist = artist[0].split('/')
    artist = artist[-2]
    artist = artist.replace('-',' ')
    Artists.append(artist)
    
print(len(Artists))
print(Artists)
```

```python
Songs = []

for a in range(len(list_url)):
    url = list_url[a]
    song = re.findall(r'(?=tab/)\S+',url)
    song = song[0].split('/')
    song = song[-1]
    song = re.sub('.(chords-\d+)$',"",song)
    song = song.replace('-',' ')
    Songs.append(song)
    
print(len(Songs))
print(Songs)
```

```python
Chords = []

for a in range(len(list_url)):
    url = list_url[a]
    
    response=requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    body = soup.find_all('div', class_='js-store')[0]
    body = str(body)
    
    chords = re.findall(r'\[ch]\S+\[/ch]',body)
    chords = [i.replace('[ch]','') for i in chords]
    chords = [i.replace('[/ch]','') for i in chords]
    chords = set(chords)
    Chords.append(chords)
    

print(len(Chords))
```

```python
print(Chords)
```

```python
df = pd.DataFrame({
    "Artists":Artists,
    "Songs":Songs,
    "Chords":Chords
})

df
```

# Data scraping

## Import libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib
import re
```

```python
url = 'http://www.ultimate-guitar.com/explore?type[]=Chords&&page=1'
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
a = soup.find_all('div', class_="js-store")[0]
a = str(a)
a = a.split('&quot;')
```

```python
list_url = []

for i in a:
    if 'https://tabs.ultimate-guitar.com/tab/' in i:
        list_url.append(i)
        
list_url
```

```python
len(list_url)
```

```python
list_url[1]
```

```python
url = list_url[1]

response=requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
b = soup.find_all('div', class_='js-store')[0]
b = str(b)

b
```

```python
chords = re.findall(r'\[ch].\[/ch]',b)
chords = [i.replace('[ch]','') for i in chords]
chords = [i.replace('[/ch]','') for i in chords]
chords = set(chords)
chords
```

```python
artist = re.findall(r'(?=tab/)\S+',url)
artist = artist[0].split('/')
artist, song = artist[1], artist[2]

artist = artist.replace('-',' ')
print(artist)

song = re.sub('.(-\d+)$',"",song)
song = song.replace('-',' ')
print(song)
```

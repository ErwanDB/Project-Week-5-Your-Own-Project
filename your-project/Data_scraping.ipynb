# Data scraping

## Import libraries

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import urllib
import re
```

## List of pages

```python
pages = []

for i in range(1,21):
    url = f'https://www.ultimate-guitar.com/explore?&type[]=Chords&order=hitstotal_desc&page={i}'
    pages.append(url)
    
pages
```

<font color=blue>***Most popular of all times = most number of hits***

## Scrap Votes & Ratings

```python
Votes = []
Ratings = []

for a in range(len(pages)):
    url = pages[a]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    body = soup.find_all('div', class_="js-store")[0]
    body = str(body)
    
    votes = re.findall(r'votes&quot;:\d+',body)
    rating = re.findall(r'rating&quot;:\S{4}',body)
    
    for b,c in zip(votes,rating):
        b = b.replace('votes&quot;:','')
        Votes.append(b)
        c = c.replace('rating&quot;:','')
        Ratings.append(c)
```

```python
print(len(Votes))
print(len(Ratings))
```

## list of url for songs

```python
list_url = []

for j in range(len(pages)):    
    url = pages[j]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    body = soup.find_all('div', class_="js-store")[0]
    body = str(body)
    body = body.split('&quot;')

    for k in body:
        if 'https://tabs.ultimate-guitar.com/tab/' in k:
            list_url.append(k)

list_url
```

```python
len(list_url)
```

## Scrap Artists

```python
Artists = []

for a in range(len(list_url)):
    url = list_url[a]
    
    artist = re.findall(r'(?=tab/)\S+',url)
    artist = artist[0].split('/')
    artist = artist[-2]
    artist = artist.replace('-',' ')
    Artists.append(artist)
    
print(len(Artists))
print(Artists)
```

## Scrap Song title

```python
Songs = []

for a in range(len(list_url)):
    url = list_url[a]
    song = re.findall(r'(?=tab/)\S+',url)
    song = song[0].split('/')
    song = song[-1]
    song = re.sub('.(chords-\d+)$',"",song)
    song = song.replace('-',' ')
    Songs.append(song)
    
print(len(Songs))
print(Songs)
```

## Scrap Chords

```python
Chords = []

for a in range(len(list_url)):
    url = list_url[a]
    
    response=requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    body = soup.find_all('div', class_='js-store')[0]
    body = str(body)
    
    chords = re.findall(r'(?=\[ch])(.*?)(?=\[/ch\])',body)
    chords = [i.replace('[ch]','') for i in chords]
    #chords = [i.replace('[/ch]','') for i in chords]
    chords = list(set(chords))
    Chords.append(chords)

print(len(Chords))
```

```python
print(Chords)
```

## Scrap Difficulty

```python
Difficulty = []

for a in range(len(list_url)):
    url = list_url[a]
    
    response=requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    body = soup.find_all('div', class_='js-store')[0]
    body = str(body)
       
    difficulty = re.findall(r'difficulty\S+has_official_version',body)
    Difficulty.append(difficulty)

print(len(Difficulty))
print(Difficulty)
```

```python
Diff = []

for i in Difficulty:
    if len(i)!=0:
        value = i[0].replace('difficulty&quot;:&quot;','')
        value = value.replace('&quot;},&quot;has_official_version','')
        value = value.replace('difficulty":"','')
        value = value.replace('"},"has_official_version','')
        Diff.append(value)
    elif len(i)==0:
        Diff.append(np.nan)
        
Diff
```

```python
len(Diff)
```

## Build DataFrame & Export

```python
df = pd.DataFrame({
    "Artists":Artists,
    "Songs":Songs,
    "Rating":Ratings,
    "Votes":Votes,
    "Difficulty":Diff,
    "Chords":Chords
})

df
```

```python
df.Difficulty.value_counts()
```

### Change Difficulty for Level

#### Define Level

```python
# easy chords without bar and neighbour frets
novice = ['A','Am','A(maj)','C','C/B','D','Dm','E','E7','Em/D','Em7','Esus','F/E','G']

# player able to navigate fretboard playing bar chords with neighbour frets
intermediate = ['Ab','Am/B','E6','F','F#7','F#sus4','F7','Fm', 'Fm7', 'Fmaj','Gb','Gm','Bm']

# same as intermediate but you learn to dampen some strings
advanced = ['Bm','Dsus2','Dsus4','Asus2','Asus4','A7','Am7','B','Bsus2','C#m','C#7','C#m7']

# complex chords with non neighbour frets and dampening strings
expert = ['A#m','A5','A6','Aadd9','Am/C','Am7/G','Amaj7/B','Asus2/E','B5','Baug','Bb','Bb/A','Bbdim','Bm/G',
          'Bm7','Bm9','Bmaj7','C#','C#5','C#M','C#dim','C2','C7/E','CM7','CM9','Cm/D#','Cm/F','Cm6/D','Cmaj7',
          'Cmaj9','D#5','D#7','D/A','D5','D6','D7/F#','D7sus4/G','D9','DM7','Dadd9','Dm7','Dmaj9','E7/G#',
          'Eb/G','EbM7','Em/B','Em/C#','Esus2','F#m/E','F/G','F6/9','FM7','Fmaj7#11','Fsus2','G#dim','G#dim7',
          'G5','G9','Gadd9','Gm/Bb','Gm6']
```

```python
def level(lst):
    
    level = []
    
    for i in lst:
        if i in expert:
            level.append("expert")
        elif i in advanced:
            level.append("advanced")
        elif i in intermediate:
            level.append("intermediate")
        elif i in novice:
            level.append("novice")
            
    if 'expert' in level:
        return "expert"
    elif 'advanced' in level and 'expert' not in level:
        return "advanced"
    elif "intermediate" in level and 'advanced' not in level and 'expert' not in level:
        return "intermediate"
    elif "novice" in level and "intermediate" not in level and 'advanced' not in level and 'expert' not in level:
        return "novice"
```

#### Add level column and drop Difficulty

```python
df["Level"]=df["Chords"].apply(level)
```
